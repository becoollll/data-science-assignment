```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data.csv")

fig, ax = plt.subplots(2, 2, figsize=(16,16))

####### 1 #######
ax[0, 0].plot(df['Township'], df['Male'], label="male", color="teal")
ax[0, 0].plot(df['Township'], df['Female'], label="female", color="mediumturquoise")
ax[0, 0].plot(df['Township'], df['Total(Male+Female)'], label="total", color="paleturquoise")
ax[0, 0].tick_params(axis='x', labelrotation = 45)
ax[0, 0].set_title("Male & Female & Total", fontsize=20)
ax[0, 0].set_xlabel("township", fontsize=18)
ax[0, 0].set_ylabel("population", fontsize=18)

ax[0, 0].legend(fontsize="12")

####### 2 #######
ax[0, 1].plot(df['Township'], df['Immigration'], label="immigration", color="#1a0")
ax[0, 1].plot(df['Township'], df['Emigration'], label="emigration", color="#1a05")
ax[0, 1].tick_params(axis='x', labelrotation = 45)
ax[0, 1].set_title("immigration & emigration", fontsize=20)
ax[0, 1].set_xlabel("township", fontsize=18)
ax[0, 1].set_ylabel("population", fontsize=18)
ax[0, 1].legend(fontsize="12")

####### 3 #######
ax[1, 0].bar(df['Township'], df['Number_of_Households'], label="Households Number", color="goldenrod")
ax[1, 0].tick_params(axis='x', labelrotation = 45)
ax[1, 0].set_title("Households Number", fontsize=20)
ax[1, 0].set_xlabel("township", fontsize=18)
ax[1, 0].set_ylabel("number", fontsize=18)
ax[1, 0].legend(fontsize="12")

####### 4 #######
ax[1, 1].plot(df['Township'], df['Birth'], label="immigration", color="mediumpurple")
ax[1, 1].plot(df['Township'], df['Death'], label="emigration", color="lavender")
ax[1, 1].tick_params(axis='x', labelrotation = 45)
ax[1, 1].set_title("Births & Deaths", fontsize=20)
ax[1, 1].set_xlabel("township", fontsize=18)
ax[1, 1].set_ylabel("population", fontsize=18)
ax[1, 1].legend(fontsize="12")

plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.35)
fig.suptitle("Miaoli County population data", fontsize=30)
plt.savefig('hw12.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data.csv")

plt.figure(figsize=(16,16))

####### 1 #######
plt.subplot(221)
plt.plot(df['Township'], df['Male'], label="male", color="teal")
plt.plot(df['Township'], df['Female'], label="female", color="mediumturquoise")
plt.plot(df['Township'], df['Total(Male+Female)'], label="total", color="paleturquoise")
plt.xticks(rotation=70)
plt.title("Male & Female & Total", fontsize=20)
plt.xlabel("township", fontsize=18)
plt.ylabel("population", fontsize=18)
plt.legend(fontsize="12")

####### 2 #######
plt.subplot(222)
plt.plot(df['Township'], df['Immigration'], label="immigration", color="#1a0")
plt.plot(df['Township'], df['Emigration'], label="emigration", color="#1a05")
plt.xticks(rotation=70)
plt.title("immigration & emigration", fontsize=20)
plt.xlabel("township", fontsize=18)
plt.ylabel("population", fontsize=18)
plt.legend(fontsize="12")

####### 3 #######
plt.subplot(223)
plt.bar(df['Township'], df['Number_of_Households'], label="Households Number", color="goldenrod")
plt.title("Households Number", fontsize=20)
plt.xticks(rotation=70)
plt.xlabel("township", fontsize=18)
plt.ylabel("number", fontsize=18)
plt.legend(fontsize="12")

####### 4 #######
plt.subplot(224)
plt.plot(df['Township'], df['Birth'], label="immigration", color="mediumpurple")
plt.plot(df['Township'], df['Death'], label="emigration", color="lavender")
plt.xticks(rotation=70)
plt.title("Births & Deaths", fontsize=20)
plt.xlabel("township", fontsize=18)
plt.ylabel("population", fontsize=18)
plt.legend(fontsize="12")

plt.suptitle("Miaoli County population data", fontsize=30)
plt.tight_layout()
plt.savefig('hw12.png')
plt.show()
```
