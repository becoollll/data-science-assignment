```python
import pandas as pd

df = pd.read_csv('data.csv')

def mean(x):
    return round(x.mean(), 4)

result = df.groupby(['High school system', 'Gender']).agg({'Num of graduates': 'sum', 
                                                           'Num of enroll in police university': 'sum', 
                                                           'Num of enroll in police college': 'sum', 
                                                           'Num of enroll in military school': ['sum', mean],}).reset_index()

result.columns = ['High school system', 
                  'Gender', 
                  'Num of graduates', 
                  'Num of enroll in police university',
                  'Num of enroll in police college', 
                  'Num of enroll in military school', 
                  'Military school enrollment average']


result['Police university rate'] = round(result['Num of enroll in police university'] / result['Num of graduates'], 4)
result['Police college rate'] = round(result['Num of enroll in police college'] / result['Num of graduates'], 4)
result['Military school enrollment rate'] = round(result['Num of enroll in military school'] / result['Num of graduates'], 4)

result = result.drop(columns='Num of enroll in military school')

result
```
